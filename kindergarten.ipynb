```python
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

```

```python
df_campaign = pd.read_csv("data/Muesli Project raw data - Campaign Data.csv")
df_orders = pd.read_csv("data/Muesli Project raw data - Orders.csv", skiprows=1) # deletes the first row
df_process = pd.read_csv("data/Muesli Project raw data - Order Process Data.csv")
df_intern = pd.read_csv("data/Muesli Project raw data - InternData Study.csv")
```

```python
df_campaign.head()
```

```python
df_campaign.info()
```

```python
df_campaign['Arrival Scan Date'] = pd.to_datetime(df_campaign['Arrival Scan Date'], format='%d/%m/%Y')
df_campaign.head()
```

```python
df_orders
```

```python
# drop column "Unnamed: 2"
df_orders.drop("Unnamed: 2", axis=1, inplace=True)

df_orders.head()
```

```python
# transforms "Order Date" from object into datetime
df_orders['Order Date'] = pd.to_datetime(df_orders['Order Date'], format='%d/%m/%Y')
df_orders.head()
```

```python
df_orders.info()
```

```python
df_process.head()
```

```python
df_process['Order Date'] = pd.to_datetime(df_process['Order Date'], format='%d/%m/%Y')
df_process['On Truck Scan Date'] = pd.to_datetime(df_process['On Truck Scan Date'], format='%d/%m/%Y')
```

```python
df_process.head()
```

```python
df_intern.head()
```

```python
df_intern['Ready to Ship Date'] = pd.to_datetime(df_intern['Ready to Ship Date'], format='%d/%m/%Y')
df_intern['Pickup Date'] = pd.to_datetime(df_intern['Pickup Date'], format='%d/%m/%Y')
```

```python
df_intern.head()
```

```python
# remove all whitespaces and replace them with "_"; make all strings in the header lower case
df_orders.columns = df_orders.columns.str.lower()
df_orders.columns = df_orders.columns.str.replace(" ", "_")
df_orders.columns = df_orders.columns.str.replace("-", "_")
df_orders.columns = df_orders.columns.str.replace("/", "_")

df_campaign.columns = df_campaign.columns.str.lower()
df_campaign.columns = df_campaign.columns.str.replace(" ", "_")

df_process.columns = df_process.columns.str.lower()
df_process.columns = df_process.columns.str.replace(" ", "_")

df_intern.columns = df_intern.columns.str.lower()
df_intern.columns = df_intern.columns.str.replace(" ", "_")
```

```python
df_intern
```

```python
df_campaign
```

```python
df_process
```

```python
df_process.drop("row_id", axis=1, inplace=True)
df_process
```

```python
df_orders.drop("unnamed:_2", axis=1, inplace=True)
df_orders
```

```python
df_orders['order_date'] = pd.to_datetime(df_orders['order_date'], format='%d/%m/%Y')
df_orders
```

```python
merge_df = pd.merge(df_orders, df_campaign, on="order_id", how="outer")
merge_df
```

```python
merge_1_df = pd.merge(df_intern, df_process, on="order_id", how="outer")
merge_1_df
```

```python
final_set = pd.merge(merge_df, merge_1_df, on="order_id", how="outer")
final_set
```

```python
# deleting the "index" column
final_set.drop("index", axis=1, inplace=True)
```

```python
final_set
```

```python
final_set['arrival_scan_date'] = pd.to_datetime(final_set['arrival_scan_date'], format='%d/%m/%Y')
```
